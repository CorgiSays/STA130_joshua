1.  The standard error of the mean and the standard deviation differ in the way in which they represent the variation among the data. In the case of the standard deviation, it is a numerical representation of the spread between datasets within that single dataset. It can show how much the individual rows of data differ from the mean of the entire dataset. On the contrary, the standard error of the mean is effectively the standard deviation of all the sample means when you bootstrap. This can be shown by bootstrapping 1000 times, and the standard error of the mean would be the variation (or standard deviation) among those means from the mean of the means.

In short, the standard deviation and the standard error of the mean capture a similar outcome and idea, but through very different methods, be it through bootstrapping then calculating the differences in the mean, attempting many experiments then calculating the differences in the means of those experiments, or any other result which result in a lot of different datasets with different means.

```python
import numpy as np # Imports numpy

original_data = np.array([52, 93, 15, 72, 61, 21, 83, 97, 75, 75]) # Creates a dataset

n_bootstrap = 10000 # sets the number of times to bootstrap

boot_medians = [np.median(np.random.choice(original_data, size=len(original_data), replace=True)) for _ in range(n_bootstrap)]
# Generates the original data 10,000 times and gathers the means into a list.

lower_bound = np.percentile(boot_medians, 2.5) # Gets the 2.5th percentile (lower bound)
upper_bound = np.percentile(boot_medians, 97.5) # Gets the 97.5th percentile (upper bound)

print(f"95% Bootstrapped Confidence Interval for the Median: ({lower_bound}, {upper_bound})")
# Prints the upper and lower bound
```

7. A confidence interval overlapping zero in this context means the following:
- At a 95% confidence interval, the point of no effect is encompassed.
- This means that in the case that there is no effect, that is within the 95% confidence interval and as such, the null hypothesis cannot be rejected, as we have a 95% confidence that the correct information (how much the drug does) falls within that range. 
A statistical confidence interval overlapping zero "fails to reject the null hypothesis" even when the sample mean statistic is shown to not be zero is thus the case, as *on average*, the data will be within the 95% confidence interval's range, and as such, it is plausible that the drug will have no effect. 

To reach the opposite conclusion, we simply need to have the 95% confidence interval fall out of that range. For example, if we had a lot of people in our sample size say that the drug did work for them, and almost nobody, or very few people say it didn't work, it is possible that after bootstrapping, the 95% confidence interval falls outside of the range in which we can "fail to reject the null hypothesis", and we can instead reject the null hypothesis which is that the drug has no effect.

To put it concisely;

We can "fail to reject the null hypothesis" (meaning to not reject the idea that the drug could be doing nothing) if the confidence interval falls within it, as we are 95% confident that the final answer falls in that range, and 0 is in that range. As a result, it is possible that the drug could do nothing, and as such we cannot reject that hypothesis.

On the contrary, we simply need to have the 95% confidence interval fall outside of the range of "no effect". This could be either via more people saying the drug didn't help them, or that it did help them. 

8. I have chosen comparison of interest to be delta Health score, effectively being represented by the initial health score minus the final health score. I chose this, as it will show the change inflicted by the vaccine. The final report of this will be shown below the following piece of code.

```python
import numpy as np
import pandas as pd
import plotly.express as px
from io import StringIO

# Function to perform bootstrapping
def bootstrap_sample(data, n_samples, random_seed):
    np.random.seed(random_seed)
    return data.sample(n=n_samples, replace=True)

# Load CSV data from a string
def load_data_from_string(csv_string):
    return pd.read_csv(StringIO(csv_string))

# Calculate the bootstrapped means and the confidence interval
def calculate_bootstrap_means(data, n_bootstrap=1000, random_seed=42):
    deltas = data['FinalHealthScore'] - data['InitialHealthScore']
    means = []

    # Create bootstrap samples and calculate their means
    for _ in range(n_bootstrap):
        sample = bootstrap_sample(data, len(data), random_seed=np.random.randint(0, 10000))
        mean_delta = (sample['FinalHealthScore'] - sample['InitialHealthScore']).mean()
        means.append(mean_delta)

    return means

# Calculate the confidence interval for the bootstrapped means
def calculate_confidence_interval(means, confidence=0.95):
    lower_bound = np.percentile(means, (1 - confidence) / 2 * 100)
    upper_bound = np.percentile(means, (1 + confidence) / 2 * 100)
    return lower_bound, upper_bound

# Main function to run the analysis
def main(csv_string, n_bootstrap=1000, random_seed=42):
    # Load the data from the string
    data = load_data_from_string(csv_string)

    # Calculate bootstrapped means
    bootstrapped_means = calculate_bootstrap_means(data, n_bootstrap, random_seed)

    # Calculate the confidence interval
    ci_lower, ci_upper = calculate_confidence_interval(bootstrapped_means)
    print(f"95% Confidence Interval for Bootstrapped Mean Delta Health Score: [{ci_lower:.2f}, {ci_upper:.2f}]")

    # Plot histogram of the bootstrapped means
    fig = px.histogram(x=bootstrapped_means, nbins=30, title='Histogram of Bootstrapped Means')
    fig.update_layout(xaxis_title='Bootstrapped Mean Delta Health Score', yaxis_title='Frequency')
    fig.add_vline(x=ci_lower, line_color='red', line_dash='dash', annotation_text='Lower CI', annotation_position='top right')
    fig.add_vline(x=ci_upper, line_color='green', line_dash='dash', annotation_text='Upper CI', annotation_position='top right')
    fig.show()

# Example CSV data as a string for testing
csv_string = """PatientID,Age,Gender,InitialHealthScore,FinalHealthScore
1,45,M,84,86
2,34,F,78,86
3,29,M,83,80
4,52,F,81,86
5,37,M,81,84
6,41,F,80,86
7,33,M,79,86
8,48,F,85,82
9,26,M,76,83
10,39,F,83,84
60,70
"""

# You can change the random seed here
main(csv_string, n_bootstrap=10000, random_seed=42)
```
