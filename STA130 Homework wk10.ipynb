1.1) The difference between simple linear regression and multiple linear regression is dependent on the number of variables/predictors they each take. In the case of single linear regression, the formula is $$y = \beta_0 + \beta_1 x + \epsilon$$, where x is the predictor and y is the variable. However, in the case of multiple linear regression, it takes multiple predictor variables to calculate a single output. The formula for multiple linear regression is $$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon$$. Independent/predictor variables are the variables we associate with change, for example, in the case of the housing market, the variable y is the output, and can be represented by something like the housing price. A variable x could be the age of the house, and potentially in the case of multiple linear regresison, there could be additional variables such as the neighbourhood safety, city, and other pieces of data. As directly mentioned before, multiple linear regression holds more benefits over single linear regression as it gives us more control over the variables and as such, we can determine more information using more variables, isolating causes for potential rises and decreasing the chance for errors.
